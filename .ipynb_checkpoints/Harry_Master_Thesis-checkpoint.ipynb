```python
#Country
        
# Setting figure size by using figure() function
fig = plt.figure(figsize = (10,5), facecolor="w")

#Background color
plt.rcParams.update({'axes.facecolor':'white'})

x = comp_df.country.unique()
y = comp_df.country.value_counts()
      
# Making the bar chart on the data
#plt.bar(x, y)

def absolute_value(val):
    a  = np.round(val/100.*y.sum(), 0)
    return a

plt.pie(y, labels=x,
        autopct=absolute_value)

plt.axis('equal')
plt.show()
```

```python
#Sub Sector

# Function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center',
                 bbox = dict(facecolor = 'black', alpha =.2))
        
# Setting figure size by using figure() function
fig = plt.figure(figsize = (12,5), facecolor="w")

#Background color
plt.rcParams.update({'axes.facecolor':'white'})

x = comp_df.comp_sector.unique()
y = comp_df.comp_sector.value_counts()
      
# Making the bar chart on the data
plt.bar(x, y)
      
# Calling the function to add value labels
addlabels(x, y)
      
# Giving title to the plot
plt.title("Number of Companies by Technology Sub-Sectors")
      
# Giving X and Y labels
plt.xlabel("Technology Sub-Sectors")
plt.ylabel("Number of Companies")


# Remove gridline
plt.grid(b=None)
      
# Visualizing the plot
plt.show()
```

```python
#Visualize missing values
missing_percent = [(master_df[series].isnull().sum()/len(master_df[series]))*100 for series in list(master_df) ]
headers_missing = [series for series in list(master_df)  ]

mean = np.mean(missing_percent)

fig, ax = plt.subplots(figsize=(15,10))
ax.set_facecolor('white')
plt.title("Missing values for each column")
plt.xlabel("Columns")
plt.ylabel("Missing values")
plt.axhline(y=mean, c='black', label='"Mean"')

ax.annotate('Mean {}'.format(round(mean,2)), xy=(0, mean), xytext=(1, 20), arrowprops=dict(facecolor='black', shrink=5),)

plt.bar(headers_missing, missing_percent)
plt.xticks(rotation=90);
```

```python
#Remove all Nans
master_nonan_df = master_df.dropna(subset=['esg_score', 'env_score', 'soc_score', 'gov_score', 'roe', 'roa', 
                                   'price_to_sales', 'debt_equity', 
                                   'value_ebitda'], how='any', axis=0)
master_nonan_df.head()
```

```python
#Groupby n1 to n5 - All years
master_n1_n5_df = master_nonan_df.groupby('ticker').agg({'esg_score': 'mean',
                                'env_score': 'mean',
                                'soc_score': 'mean',
                                'gov_score': 'mean',
                                'roe': 'median',
                                'roa': 'median',
                                'price_to_sales': 'median',
                                'value_ebitda': 'median',
                                't_assets': 'mean'}).reset_index() 
```

```python
#Rename columns
master_n1_n5_df.rename(columns={'esg_score': 'esg_score_n1_n5',
                         'env_score': 'env_score_n1_n5',
                         'soc_score': 'soc_score_n1_n5',
                         'gov_score': 'gov_score_n1_n5',
                         'roe': 'roe_n1_n5',
                         'roa': 'roa_n1_n5',
                         'price_to_sales': 'price_to_sales_n1_n5',
                         'value_ebitda': 'value_ebitda_n1_n5',
                         't_assets': 't_assets_n1_n5'},inplace=True)
```

```python
#Groupby n1 to n3 and n4 to n5 - Pre and Post Covid

#Pre Covid filter
master_n1_n3_df = master_nonan_df[master_nonan_df['year']<2020]

#Groupby
master_n1_n3_agg_df = master_n1_n3_df.groupby('ticker').agg({'esg_score': 'mean',
                                'env_score': 'mean',
                                'soc_score': 'mean',
                                'gov_score': 'mean',
                                'roe': 'median',
                                'roa': 'median',
                                'price_to_sales': 'median',
                                'value_ebitda': 'median',
                                't_assets': 'mean'}).reset_index() 

#Rename columns
master_n1_n3_agg_df.rename(columns={'esg_score': 'esg_score_n1_n3',
                         'env_score': 'env_score_n1_n3',
                         'soc_score': 'soc_score_n1_n3',
                         'gov_score': 'gov_score_n1_n3',
                         'roe': 'roe_n1_n3',
                         'roa': 'roa_n1_n3',
                         'price_to_sales': 'price_to_sales_n1_n3',
                         'value_ebitda': 'value_ebitda_n1_n3',
                         't_assets': 't_assets_n1_n3'},inplace=True)

#Post Covid filter
master_n4_n5_df = master_nonan_df[master_nonan_df['year']>2019]

#Groupby
master_n4_n5_agg_df = master_n4_n5_df.groupby('ticker').agg({'esg_score': 'mean',
                                'env_score': 'mean',
                                'soc_score': 'mean',
                                'gov_score': 'mean',
                                'roe': 'median',
                                'roa': 'median',
                                'price_to_sales': 'median',
                                'value_ebitda': 'median',
                                't_assets': 'mean'}).reset_index() 
#Rename columns
master_n4_n5_agg_df.rename(columns={'esg_score': 'esg_score_n4_n5',
                         'env_score': 'env_score_n4_n5',
                         'soc_score': 'soc_score_n4_n5',
                         'gov_score': 'gov_score_n4_n5',
                         'roe': 'roe_n4_n5',
                         'roa': 'roa_n4_n5',
                         'price_to_sales': 'price_to_sales_n4_n5',
                         'value_ebitda': 'value_ebitda_n4_n5',
                         't_assets': 't_assets_n4_n5'},inplace=True)
```

```python
#Outliers
def WinsorizeCustom(data):
    quantiles = data.quantile([0.05, 0.95])
    q_05 = quantiles.loc[0.05]
    q_95 = quantiles.loc[0.95]

    out = np.where(data.values <= q_05,q_05, 
                                      np.where(data >= q_95, q_95, data)
                  )
    return out

master_n1_n5_df['roe_n1_n5_no_outliers'] = WinsorizeCustom(master_n1_n5_df['roe_n1_n5'])
```
